```python

def single_patch(data,h0,h1,w0,w1,d0,d1):
  chunk= np.zeros((60,60,60,4))
  print('data',data.shape)
  chunk[:,:,:,:]=data[h0:h1,w0:w1,d0:d1,:] 
  reshaped_chunk=chunk.reshape(60,60,60,4)
  plt.imshow(reshaped_chunk[:,:,45,1])
  plt.show()
  # reshaped_chunk=model_predict_patch(reshaped_chunk)
  return reshaped_chunk
def create_patches(data):
    predicted_volume=np.zeros((240,240,155,4))
    print(data.shape)
    for u in range(1,3):
      for y in range(1,5):
        for z in range(1,5):
          print(' start x-axis=',(y-1)*60,' end= x-axis',(y)*60, ' start y-axis=',(z-1)*60,' end= y-axis',(z)*60,' start z-axis=',((u-1)*60)+13,' end= z-axis',((u)*60)+13)
          predicted_slice=single_patch(data,(y-1)*60,(y)*60,(z-1)*60,(z)*60,((u-1)*60)+13,((u)*60)+13)
          print('predicted',predicted_slice.shape)
          predicted_volume[((y-1)*60):((y)*60),((z-1)*60):((z)*60),(((u-1)*60)+13):(((u)*60)+13)]=predicted_slice
    plt.imshow(predicted_volume[:,:,70,1])
    plt.show()
create_patches(data[:,:,:,:])
```
